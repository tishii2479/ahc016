```python
with open("../data/n_map.txt", "r") as f:
    s = f.readline()
```

```python
import json
json_str = s.replace("'", "").replace("(", "\"").replace(")", "\"").replace(" ", "")
json_data = json.loads(json_str)
```

```python
json_data
```

```python
best_score = [[0] * 41 for _ in range(91)]
n_map = [[0] * 41 for _ in range(91)]

for param, score in json_data.items():
    m, eps, n = param.split(",")
    m = int(m)
    eps = float(eps)
    n = int(n)
    m_idx = m - 10
    eps_idx = round(eps * 100)

    if score > best_score[m_idx][eps_idx]:
        best_score[m_idx][eps_idx] = score
        n_map[m_idx][eps_idx] = n
```

```python
for i in range(91):
    for j in range(1, 41):
        n_map[i][j] = max(n_map[i][j], n_map[i][j - 1])

for i in range(1, 91):
    for j in range(41):
        n_map[i][j] = max(n_map[i - 1][j], n_map[i][j])
```

```python
with open("../data/rs_n_map.txt", "w") as f:
    f.write(str(n_map))
```
