```python
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
```

```python
def f(s):
    scores = []
    for line in s.strip().split("\n"):
        scores.append([])
        for e in line.strip().split():
            scores[-1].append(float(e))

    return scores

def score_f(a, b):
    scores_a = f(a)
    scores_b = f(b)

    scores_diff = [[0] * len(scores_a[0]) for _ in range(len(scores_a))]
    for i in range(len(scores_a)):
        for j in range(len(scores_a[i])):
            scores_diff[i][j] = (scores_a[i][j] - scores_b[i][j]) / scores_b[i][j] * 100
            print(f"{scores_diff[i][j]:8.3f}", end=" ")
        print()

    return scores_diff

def err_f(a, b):
    err_a = f(a)
    err_b = f(b)

    err_diff = [[0] * len(err_a[0]) for _ in range(len(err_a))]
    for i in range(len(err_a)):
        for j in range(len(err_a[i])):
            err_diff[i][j] = err_a[i][j] - err_b[i][j]
            print(f"{err_diff[i][j]:8.3f}", end=" ")
        print()

    return err_diff
```
